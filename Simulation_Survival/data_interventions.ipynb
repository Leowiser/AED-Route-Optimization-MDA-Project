```python
# filter on type of intervention
cardiac_arrest = ['P003 - HARTSTILSTAND - DOOD - OVERLEDEN']
cad_9_cardiac = filtered_cad_9[filtered_cad_9['EventType Trip'].isin(cardiac_arrest)]
```

```python
# drop rows with None values
# Remove rows with 'None' values only from specific columns
cad_9_cardiac.dropna(subset=['T1', 'T3'], inplace=True)
```

```python
cad_9_cardiac
```

```python
# can only use the ones that have T1, T3, longitude and latitude filer out the rest 
filtered_cad_9_cardiac = cad_9_cardiac[cad_9_cardiac['T1'].notna()]
filtered_cad_9_cardiac = cad_9_cardiac[cad_9_cardiac['T3'].notna()]
filtered_cad_9_cardiac = cad_9_cardiac[cad_9_cardiac['Longitude intervention'].notna()]
filtered_cad_9_cardiac = cad_9_cardiac[cad_9_cardiac['Latitude intervention'].notna()]
filtered_cad_9_cardiac
```

```python
# filter data, want to save T2, T3, latitude + longitude 
final_interventions = filtered_cad_9_cardiac[['T1', 'T3', 'Latitude intervention', 'Longitude intervention']]
```

```python
final_interventions
```

```python
# reset index 
final_interventions = final_interventions.reset_index(drop=True)
final_interventions
```

```python
# make a column for dif between T2 and T3

# calculate the delivertime 
# Convert date columns to datetime objects
final_interventions ['T1'] = pd.to_datetime(final_interventions ['T1'], format='%d%b%y:%H:%M:%S')
final_interventions ['T3'] = pd.to_datetime(final_interventions ['T3'])

# Calculate delivery time
final_interventions['Response Time'] = (final_interventions ['T3'] - final_interventions ['T1']).dt.total_seconds()
```

```python
final_interventions
```

```python
# there are some interventions that appear twice, same patient spot and T3 but different T1, probably sveral calls for same incidence
# remove these by removing rows that have exactly the same latitude + longitude + T3 and keep the first one 
# the first one is the first call
# Drop duplicates based on 'T3', 'Longitude intervention', and 'Latitude intervention' columns
final_interventions = final_interventions.drop_duplicates(subset=['T3', 'Longitude intervention', 'Latitude intervention'])
```

```python
final_interventions
```

```python
# reset index 
final_interventions = final_interventions.reset_index(drop=True)
final_interventions
```

```python
# save the intervantions dataset 

file_path = "Data/interventions.csv"

# Save the dataframe to a CSV file
final_interventions.to_csv(file_path, index=False)

# If you want to include the index in the CSV file, remove index=False
# df.to_csv(file_path)

print("DataFrame saved to CSV successfully!")
```
